# Simulator

```python
simulator = GaussianMean(param_grid_bounds=(-10, 10), 
                         param_grid_size=1000, 
                         likelihood_kwargs={'scale': 1}, 
                         prior='gaussian', prior_kwargs={'loc': 0, 'scale': 2}, 
                         observed_sample_size=1, sample_method='mle', param_dims=1, observed_dims=1)
```

### Simulated data

```python
b_sample_theta, b_sample_x = simulator.generate_sample(size=20_000)
b_prime_sample_theta, b_prime_sample_x = simulator.generate_sample(size=20_000, params=np.linspace(-13, 13, 20_000))
```

### Observed data

```python
observed_theta, observed_x = simulator.generate_sample(size=10_000, params=np.linspace(-7, 7, 10_000))
```

# Wald

```python
wald_confidence_sets = []
chi2_cutoffs = np.repeat(chi2(df=simulator.d).ppf(q=CONFIDENCE_LEVEL), len(simulator.param_grid))
for obs_x in tqdm(observed_x):
    wald_statistics = simulator.compute_wald(sample=obs_x, parameter=simulator.param_grid).reshape(-1, )
    assert wald_statistics.shape == chi2_cutoffs.shape
    wald_confidence_sets.append(simulator.param_grid[wald_statistics <= chi2_cutoffs])
```

# Prediction Intervals

```python
posterior_means = simulator.exact_cond_mean(sample=observed_x)
posterior_vars = simulator.exact_cond_var()

if CONFIDENCE_LEVEL == 0.95:
    z_alpha2 = 1.96
else:
    raise NotImplementedError
    
prediction_sets = np.hstack(( (posterior_means - z_alpha2*np.sqrt(posterior_vars)).reshape(-1, 1), (posterior_means + z_alpha2*np.sqrt(posterior_vars)).reshape(-1, 1)))
```

```python
wald_coverage = np.zeros(simulator.param_grid.shape)
prediction_coverage = np.zeros(simulator.param_grid.shape)
waldo_coverage = np.zeros(simulator.param_grid.shape)

for idx, theta in tqdm(enumerate(simulator.param_grid)):
    p, data = simulator.generate_sample(size=10_000, params=theta)
    
    # WALD
    wald_stats = simulator.compute_wald(sample=data[::10], parameter=theta).reshape(-1, )
    wald_cutoffs = np.repeat(chi2(df=simulator.d).ppf(q=CONFIDENCE_LEVEL), len(data[::10]))
    wald_coverage[idx] = np.mean(wald_stats <= wald_cutoffs)
    
    # PREDICTIONS
    if CONFIDENCE_LEVEL == 0.95:
        z_alpha2 = 1.96
    else:
        raise NotImplementedError
    posterior_means = simulator.exact_cond_mean(sample=data)
    posterior_vars = simulator.exact_cond_var()
    prediction_coverage[idx] = np.mean( (theta <= (posterior_means+z_alpha2*np.sqrt(posterior_vars))) & (theta >= (posterior_means-z_alpha2*np.sqrt(posterior_vars))))
    
    # WALDO
    obs_cond_means, _ = conditional_mean.fit(data_predict=data.reshape(-1, simulator.observed_d))
    obs_cond_vars, _ = conditional_var.fit(data_predict=data.reshape(-1, simulator.observed_d))
    waldo_stats = (np.subtract(obs_cond_means.reshape(-1,1), np.array([theta]).reshape(-1, 1))**2)/obs_cond_vars.reshape(-1, 1)
    waldo_coverage[idx] = np.mean(waldo_stats.reshape(-1, ) <= qr_XGB.predict(np.array([theta]).reshape(-1, 1)))
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 15))

ax.plot(simulator.param_grid, wald_coverage, color="crimson", linewidth=5)
ax.plot(simulator.param_grid, prediction_coverage, color="darkgreen", linewidth=5, linestyle='-.')
ax.plot(simulator.param_grid, waldo_coverage, color="orange", linewidth=7, linestyle='--')
ax.plot(simulator.param_grid, np.repeat(CONFIDENCE_LEVEL, simulator.param_grid.shape[0]), label="Nominal coverage = 95%", color="black", linewidth=9, linestyle=(0, (1, 1)), zorder=4)
ax.set_xlabel(r"$\theta$", fontsize=45)
ax.set_ylabel("Coverage", fontsize=45)
ax.tick_params(axis='both', which='major', labelsize=30, bottom=True, left=True, labelleft=True, labelbottom=True)
ax.set_title("Empirical Coverage", fontsize=45, pad=12)
# ax.set_ylim(0.5, 1)
ax.legend(loc=(0.15, 0.01), prop={'size': 40})

plt.savefig(os.path.join(PATH, 'waldo/results/coverage_analysis/coverage.pdf'), bbox_inches="tight")
plt.show()
```
